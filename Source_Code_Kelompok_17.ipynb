```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('train.csv')



sns.set_style("whitegrid")
```

```python
df = df.dropna()

print(df.isnull().sum())
```

```python
duplicates = df.duplicated().sum()
print(f"Jumlah duplikasi: {duplicates}")

df = df.drop_duplicates()

duplicates = df.duplicated().sum()
print(f"Jumlah duplikasi setelah penghapusan: {duplicates}")
```

```python
# Data cleaning
df = df.drop(columns=['Unnamed: 0', 'id']) 
df.rename(columns={'satisfaction': 'Satisfaction'}, inplace=True)  
df['Satisfaction'] = df['Satisfaction'].replace({'neutral or dissatisfied': 'dissatisfied'})
df['Satisfaction_encoded'] = df['Satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)
```

```python
# 1. Chart Donut untuk distribusi kepuasan penumpang
plt.figure(figsize=(6, 6))
palette = {'dissatisfied': '#3496f7', 'satisfied': '#fa3737'}
satisfaction_counts = df['Satisfaction'].value_counts()
plt.pie(satisfaction_counts, labels=None, 
        autopct='%1.1f%%', colors=palette.values(), 
        startangle=270, wedgeprops={'width': 0.4, 'edgecolor': 'white'}, 
        shadow=True, pctdistance=0.8, textprops={'fontsize': 15, 'color': 'white'})
plt.legend(satisfaction_counts.index, loc='upper right', fontsize=12)

plt.title('Overall Passenger Satisfaction Distribution', fontsize=16, pad=20)
plt.tight_layout()
plt.show()
```

```python
# 2 . Histogram untuk distribusi usia penumpang
plt.figure(figsize=(8, 6))
sns.histplot(data=df, x='Age', bins=20, kde=True, color='#3496f7', line_kws={'color': 'red'})
plt.title('Age Distribution of Passengers', fontsize=16, pad=20)
plt.xlabel('Age', fontsize=14)
plt.ylabel('Frequency', fontsize=14)
plt.show()
```

```python
# 3. Distribution kepuasan penumpang berdasarkan umur
plt.figure(figsize=(10, 6))
sns.violinplot(x='Satisfaction', y='Age', data=df, palette=palette, inner='quartile')
plt.title('Age Distribution by Satisfaction Level', fontsize=14)
plt.xlabel('Satisfaction Level')
plt.ylabel('Age')
plt.tight_layout()
plt.show()
```

```python
# 4. Perbandingan Kepuasan antar Kelas Penerbangan (dengan Persentase dan Urutan)

satisfaction_order = ['dissatisfied', 'satisfied']

class_total = df['Class'].value_counts()

plt.figure(figsize=(8,5))
ax = sns.countplot(x='Class', hue='Satisfaction', data=df, palette=palette, hue_order=satisfaction_order)

for container in ax.containers:
    for bar in container:
        height = bar.get_height()
        class_name = bar.get_x() + bar.get_width()/2
        class_idx = int(round(class_name))
        class_label = ax.get_xticklabels()[class_idx].get_text()
        total = class_total[class_label]
        percent = 100 * height / total if total != 0 else 0
        ax.text(bar.get_x() + bar.get_width()/2, height + 5, f'{percent:.1f}%', ha='center', va='bottom', fontsize=9)

plt.title('Comparison of Satisfaction Based on Flight Class', fontsize=16)
plt.xlabel('Flight Class')
plt.ylabel('Number of Passengers')
plt.legend(title='Satisfaction Level', loc='upper left')
plt.tight_layout()
plt.show()
```

```python
# 5. Violin plot untuk distribusi jarak penerbangan berdasarkan kelas dan kepuasan

plt.figure(figsize=(10,6))
sns.violinplot(x='Class', y='Flight Distance', hue='Satisfaction', data=df, split=True,palette=palette)
plt.title('Distribusi Flight Distance berdasarkan Kelas dan Kepuasan')
plt.xlabel('Class')
plt.ylabel('Flight Distance')
plt.legend(title='Satisfaction')
plt.show()
```

```python
# 6. Pie Chart untuk kepuasan penumpang berdasarkan keterlambatan penerbangan
colors = ['#3496f7', '#fa3737']  

delay_df = df[(df['Departure Delay in Minutes'] < 120) & 
              (df['Arrival Delay in Minutes'] < 120)].copy()

delay_df['Delay Group'] = pd.cut(
    delay_df['Departure Delay in Minutes'],
    bins=[-1, 15, 30, 60, 120],
    labels=['0-15', '16-30', '31-60', '61-120']
)

delay_groups = ['0-15', '16-30', '31-60', '61-120']

fig, axes = plt.subplots(1, 4, figsize=(18, 6))  # 1 row, 4 columns

for i, delay_group in enumerate(delay_groups):
    group_data = delay_df[delay_df['Delay Group'] == delay_group]
    satisfaction_counts = group_data['Satisfaction'].value_counts(normalize=True).reindex(['satisfied', 'dissatisfied'], fill_value=0)
    sizes = [satisfaction_counts['satisfied'] * 100, satisfaction_counts['dissatisfied'] * 100]
    labels = ['Satisfied', 'Dissatisfied']
    ax = axes[i]  
    ax.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=270, colors=colors, explode=(0.1, 0), shadow=True,textprops={'fontsize': 13, 'color': 'white'})
    ax.set_title(f'Passenger Satisfaction: {delay_group} Minutes Delay', fontsize=13)

plt.suptitle('Passenger Satisfaction Based On Flight Delay', fontsize=18, y=1)
plt.legend(satisfaction_counts.index, bbox_to_anchor=(1, 1.5), fontsize=12)

plt.tight_layout()
plt.show()
```

```python
# 7. Linear regression untuk hubungan antara keterlambatan keberangkatan dan kepuasan penumpang
delay_df['Satisfaction_Encoded'] = delay_df['Satisfaction'].map({'satisfied': 1, 'dissatisfied': 0})

plt.figure(figsize=(10,6))
sns.regplot(
    data=delay_df,
    x='Departure Delay in Minutes',
    y='Satisfaction_Encoded',
    logistic=False, 
    scatter_kws={'alpha':0.002},
    line_kws={'color': 'red'}
)
plt.title('Linear Relationship: Departure Delay vs Passenger Satisfaction')
plt.xlabel('Departure Delay (Minutes)')
plt.ylabel('Probability of Satisfaction')
plt.grid(True)
plt.show()
```

```python
# 8 . Jointplot untuk hubungan antara umur dan jarak penerbangan
sns.jointplot(x='Age', y='Flight Distance', data=df, kind='scatter', color='#3496f7',alpha=0.025)
plt.suptitle('Jointplot: Age vs Flight Distance', fontsize=16, y=1.1)
plt.show()
```

```python
# 9 . Barplot untuk hubungan antara fitur layanan dan kepuasan penumpang
df['satisfaction_encoded'] = df['Satisfaction'].apply(lambda x: 1 if x == 'satisfied' else 0)

service_cols = ['Inflight wifi service', 'Ease of Online booking',
                'Gate location', 'Food and drink', 'Online boarding', 'Seat comfort', 
                'Inflight entertainment', 'On-board service', 'Leg room service', 
                'Baggage handling', 'Checkin service', 'Inflight service', 'Cleanliness']

correlations = df[service_cols + ['satisfaction_encoded']].corr()['satisfaction_encoded'].drop('satisfaction_encoded')

correlations_sorted = correlations.sort_values(ascending=False)

plt.figure(figsize=(15,6))
sns.barplot(x=correlations_sorted.values, y=correlations_sorted.index, palette='coolwarm')

for index, value in enumerate(correlations_sorted.values):
    plt.text(value + 0.01, index, f"{value:.2f}", va='center')
    
plt.title('Correlation of Service Features with Satisfaction', fontsize=16)
plt.xlabel('Correlation Value')
plt.ylabel('Service Features')

plt.grid(axis='x', linestyle='--', alpha=0.7)
plt.xlim(0, 0.6)  
plt.tight_layout()
plt.show()
```
